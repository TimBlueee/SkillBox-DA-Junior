```python
import pandas as pd 
import numpy as np
import json
import glob
import os
```

```python
df_native = pd.read_csv('копия native.csv')
df_native = df_native.drop(columns=['Unnamed: 0'])
df_native['ad_channel'] = 'native'
df_native['price'] = 2.00
df_native.head()
```

```python
df_banner = pd.read_excel("копия banner.xlsx")
df_banner['ad_channel'] = 'banner'
df_banner['price'] = 0.50
df_banner.head()
```

```python
df_traffic = pd.read_csv('копия organic_traffic.csv')
df_traffic = df_traffic.rename(columns = {'interest': 'funnel_stage','150001': 'timestamp', 'Unnamed: 4' : 'profit', '764740' : 'user_id', 'iOS': 'os'})
df_traffic = df_traffic.drop(columns=['17'])
df_traffic['ad_channel'] = 'traffic'
df_traffic['price'] = 0.00
df_traffic.head()
```

```python
df_network = pd.read_csv('копия partner_network.csv')
df_network = df_network.rename(columns={'iOS': 'os', '150000': 'timestamp', 'interest': 'funnel_stage', 'Unnamed: 4' : 'profit', '561010' : 'user_id'})
df_network = df_network.drop(columns=['8'])
df_network['ad_channel'] = 'partner_network'
df_network['price'] = 1.00
df_network.head()
```

```python
df_rich = pd.read_excel("копия rich.xlsx")
df_rich['funnel_stage'] = df_rich['funnel_stage'].str.lower()
df_rich['ad_channel'] = 'rich'
df_rich['price'] = 2.50
df_rich.head()
```

```python
df_video = pd.read_csv('копия video.csv')
df_video = df_video.drop(columns=['Unnamed: 0'])
df_video['ad_channel'] = 'video'
df_video['price'] = 0.75
df_video.head()
```

```python
df_other = pd.read_csv('копия other.csv')
df_other = df_other.drop(columns=['Unnamed: 0'])
df_other['ad_channel'] = 'other'
df_other['price'] = 'NaN'
df_other.head()
```

```python
df_adv = pd.read_json('Копия advertising_information.json')
df_adv
```

```python
df_data = pd.concat([df_native, df_banner, df_traffic, df_network, df_rich, df_video, df_other], axis = 0)
df_data = df_data.sort_values('timestamp')
df_data
```

```python
df_data.to_csv('data.csv', encoding='utf-8')
```
